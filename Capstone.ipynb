```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('uk_renewable_energy.csv')
```

```python
df ###Mtoe(megatonne of oil equivalent)

##For context 1 mtoe = 42 petajoules (42 x 10^15 J)
## 1 pj = 0.2778 tWh
##42 pj = 11.67 tWh
```

```python
df['(%)Renewable_and_waste'] = df['(%)Renewable_and_waste'].apply(lambda x: x * 100)
```

```python
df.corr()
```

```python
plt.figure(figsize = (25, 11))
sns.heatmap(df.corr(),annot=True, mask=np.triu(df.corr()))
plt.show()
```

```python
total_vs_renewable = df[['Year', 'Renewable_and_waste_energy', 'Total_energy_consumption_primary_fuels_and_equivalents']]
```

```python
total_vs_renewable
```

```python
total_vs_renewable.corr()
```

```python
sns.pairplot(data=total_vs_renewable)
```

```python
# Set color palette for color-blind friendly colors
color_palette = sns.color_palette("colorblind")
sns.set_palette(color_palette)
# Create a pair plot
sns.pairplot(data=total_vs_renewable)

# Create line plots
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
renewable_line = sns.lineplot(data=total_vs_renewable, x='Year', y='Renewable_and_waste_energy', label='Renewable Energy')
total_line = sns.lineplot(data=total_vs_renewable, x='Year', y='Total_energy_consumption_primary_fuels_and_equivalents', label='Total Energy Consumption')
plt.title('Difference in Energy Consumption')
plt.xlabel('Year')
plt.ylabel('Megatonne of oil equivalent (mtoe)')
plt.legend()  # Add a legend to distinguish lines

# Calculate the midpoint between the two lines
midpoint_x = total_vs_renewable['Year'].iloc[-1]  # The last year
midpoint_y = (total_vs_renewable['Renewable_and_waste_energy'].iloc[-1] +
            total_vs_renewable['Total_energy_consumption_primary_fuels_and_equivalents'].iloc[-1]) / 2

# Add an annotation with an arrow connecting the midpoint to both lines
plt.annotate('2020: 14.4% renewable energy', xy=(midpoint_x, midpoint_y), xytext=(midpoint_x - 5, midpoint_y + 10),
             arrowprops=dict(facecolor='white', arrowstyle='->'),
             fontsize=12, horizontalalignment='center')  # Annotate the 2020 point

# Add text labels at the end of the lines
plt.text(total_vs_renewable['Year'].iloc[-1], total_vs_renewable['Total_energy_consumption_primary_fuels_and_equivalents'].iloc[-1] -5,
         f"2020 Total: {total_vs_renewable['Total_energy_consumption_primary_fuels_and_equivalents'].iloc[-1]:.2f} mtoe", ha='right')
plt.text(total_vs_renewable['Year'].iloc[-1], total_vs_renewable['Renewable_and_waste_energy'].iloc[-1] + 10,
         f"2020 Renewable: {total_vs_renewable['Renewable_and_waste_energy'].iloc[-1]:.2f} mtoe", ha='right')

plt.grid(True)  # Add grid lines to the plot
plt.show()

```

```python
sns.lineplot(data=total_vs_renewable, x='Year', y='Renewable_and_waste_energy')
sns.lineplot(data=total_vs_renewable, x='Year', y='Total_energy_consumption_primary_fuels_and_equivalents')
plt.title('Difference')
plt.ylabel('Megatonne of oil equivalent(mtoe)')
plt.show()

##2020 only 14.4% of renewable energy
```

```python

# Reshape the DataFrame using pd.melt()
melted_df = pd.melt(df_uk_clean, id_vars=['year'], value_vars=['electricity_demand_tWh', 'electricity_generation_tWh',                                               'renewables_electricity_tWh', 'nuclear_electricity_tWh'],
                    var_name='variable', value_name='value')

# Create the bar plot
plt.figure(figsize=(20, 20))
sns.set_palette("colorblind")
sns.barplot(data=melted_df[['year', 'value', 'variable']][melted_df['variable'].isin(['electricity_demand_tWh', 'electricity_generation_tWh'])],
            x='year', y='value', hue='variable', errorbar=None)

plt.title('Demand and Supply', fontsize=18)  # Increase title font size
plt.xlabel('Year', fontsize=14)  # Increase x-axis label font size
plt.ylabel('tWh', fontsize=14)  # Increase y-axis label font size
plt.xticks(fontsize=14)  # Increase x-axis tick label font size
plt.yticks(fontsize=14)  # Increase y-axis tick label font size
plt.legend(title='Energy Type', loc='upper left', fontsize=14)
plt.show()
```
